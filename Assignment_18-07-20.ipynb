```python
from sklearn.preprocessing import scale
nor_df=scale(df.iloc[:,1:])
nor_df
```

```python
from scipy.cluster.hierarchy import linkage
a=linkage(nor_df,method="complete",metric="euclidean")
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(30,30))
sch.dendrogram(a)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
clust = AgglomerativeClustering(n_clusters = 5, linkage='complete',affinity='euclidean').fit(nor_df)
```

```python
clust.labels_
```

```python
cluster=pd.DataFrame(clust.labels_)
cluster
```

```python
df['cluster']=cluster
df.head()
```

# K-Means Non-Heirarachial Clustering

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1, 6):
    kmeans = KMeans(n_clusters=i, init= 'k-means++', random_state=0)
    kmeans.fit(nor_df)
    wcss.append(kmeans.inertia_)

wcss
```

```python
plt.plot(range(1,6), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Here k=2 is the best for Clustering
